```python
import pandas as pd
from pandasai import SmartDataframe
from pandasai.llm.local_llm import LocalLLM
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:,.0f}'.format
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv("sales.csv")
df.head()
```

```python
import pygwalker as pyg
```

```python
pyg.walk(df)
```

```python
df.chat("Quel est le total des ventes ?")
```

```python
df.chat('Quelle est la vente par catégorie de produit ?')
```

```python
ollama_llm = LocalLLM(api_base="http://localhost:11434/v1", model="llama3")
df_llm = SmartDataframe(df, config={"llm": ollama_llm})
```

```python
df_llm.chat('Quel est le total des ventes ?')
```

```python
df_llm.chat('Quelle est la vente par catégorie de produit ?')
```

```python
df_llm.chat('Afficher les ventes par catégorie de produit sur un graphique')
```

```python
df_llm.chat('Quelles catégories de produits ont les meilleures et les pires ventes ?')
```

```python
df_llm.chat('Comment les ventes par catégorie de produit ont-elles évolué au fil des mois ou des années ?')
```
